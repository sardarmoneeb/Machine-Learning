```python
sns.distplot(["installment"])
plt.show()
```

```python
min_range = dataset["installment"].mean()-(3*dataset["installment"].std())
max_range = dataset["installment"].mean()+(3*dataset["installment"].std())
```

```python
n_dataset=dataset[dataset["installment"]<=max_range]
```

```python
sns.boxplot(x="installment",data=dataset)
plt.show()
```

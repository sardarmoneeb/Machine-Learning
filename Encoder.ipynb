```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size":["s","m","l","xl","m","s","m"]})
```

```python
df.head(4)
```

# Ordinal Encoding through sklearn

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord_data = [["s", "m", "l","xl"]] 
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
df["size_en"]=oe.fit_transform(df[["Size"]])
```

```python
df
```

# Ordinal Encoding through map function

```python
ord_data1 = {"s":0,"m":2,"l":3,"xl":4}
```

```python
df["Size_en_map"]=df["Size"].map(ord_data1)
```

```python
df
```

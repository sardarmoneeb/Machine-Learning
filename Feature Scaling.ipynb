```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel(r"D:\Kaggle datasets\loan_dataset.xlsx")
```

```python
dataset.head(5)
```

```python
dataset.describe()
```

```python
sns.boxplot(x="loan_amount",data=dataset)
plt.show()
```

# Scaling through Standard Scalar

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["loan_amount"]])
```

```python
dataset["loan_amount_ss"]=pd.DataFrame(ss.transform(dataset[["loan_amount"]]),columns=["x"])
```

```python
dataset.head(10)
```

```python
sns.distplot(dataset["annual_inc"])
plt.show()
```

# Scaling through Normalization("min,max scaler")

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["annual_inc"]])
```

```python
dataset["annual_inc_ms"]=ms.transform(dataset[["annual_inc"]])
```

```python
dataset.head(10)
```

```python
sns.distplot(dataset["annual_inc_ms"])
plt.show()
```
